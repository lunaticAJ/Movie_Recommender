```python
import numpy as np
import pandas as pd
```

```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2020"
d1 = pd.read_html(link, header = None)[2]
d2 = pd.read_html(link, header = None)[3]
d3 = pd.read_html(link, header = None)[4]
d4 = pd.read_html(link, header = None)[5]

d = pd.concat([d1,d2,d3,d4],ignore_index = True)
d
```

```python
for i in range(d.shape[0]):
    a = 0
    for j in range(len(d["Title"][i])):
        if (d["Title"][i][j] == '('):
            a = j
    if a != 0:
        d["Title"][i] = d["Title"][i][:a]
```

```python
from tmdbv3api import TMDb
from tmdbv3api import Movie
import json
import requests
tmdb = TMDb()
tmdb_movie = Movie()
tmdb.api_key = 'b911be8c78ffa201769644bdeb318d41'
```

```python
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    if data_json['genres']:
        genre_str = " " 
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        genre_list = str(genres)
        return genre_list
    else:
        np.NaN
```

```python
d['Genre'] = d['Title'].map(lambda x: get_genre(str(x)))
```

```python
def list_to_string(x):
    if x:
        return x.replace(',', '').replace('[','').replace(']','').replace("'",'')
    else:
        return np.NaN

d['Genre'] = d['Genre'].map(lambda x: list_to_string(x))
```

```python
d
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]

def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])

def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
d['director_name'] = d['Cast and crew'].map(lambda x: get_director(str(x)))
d['actor_1_name'] = d['Cast and crew'].map(lambda x: get_actor1(str(x)))
d['actor_2_name'] = d['Cast and crew'].map(lambda x: get_actor2(str(x)))
d['actor_3_name'] = d['Cast and crew'].map(lambda x: get_actor3(str(x)))
```

```python
df = d.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','Genre','Title']]
df.fillna('unknown', inplace=True)
```
