```python
import numpy as np
import pandas as pd
```

```python
credits = pd.read_csv('/content/drive/MyDrive/Movie Datasets/credits.csv/credits.csv')
meta = pd.read_csv('/content/drive/MyDrive/Movie Datasets/movies_metadata.csv/movies_metadata.csv')
credits
```

```python
meta
```

```python
meta['year'] = (pd.to_datetime(meta['release_date'], errors='coerce')).dt.year
```

```python
meta
```

```python
meta = meta.loc[meta.year <= 2017,['genres','id','title','year']]
meta = meta.loc[meta.year >= 2015,['genres','id','title','year']]
meta['id'] = meta['id'].astype(int)
meta['year'] = meta['year'].astype(int)
meta
```

```python
data17 = pd.merge(meta, credits, on='id')
data17
```

```python
import ast
data17['genres'] = data17['genres'].map(lambda x: ast.literal_eval(x))
data17['cast'] = data17['cast'].map(lambda x: ast.literal_eval(x))
data17['crew'] = data17['crew'].map(lambda x: ast.literal_eval(x))
```

```python
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))

data17['genres'] = data17['genres'].map(lambda x: make_genresList(x))
data17
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])


def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])


def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])

def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data17['actor_1_name'] = data17['cast'].map(lambda x: get_actor1(x))
data17['actor_2_name'] = data17['cast'].map(lambda x: get_actor2(x))
data17['actor_3_name'] = data17['cast'].map(lambda x: get_actor3(x))
data17['director_name'] = data17['crew'].map(lambda x: get_directors(x))
```

```python
data17
```

```python
data17 = data17.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','title']]
```

```python
data17
```

```python
data17 = data17.dropna(how='any')
data17 = data17.rename(columns={'genres':'Genre'})
data17 = data17.rename(columns={'title':'Title'})
```

```python
data17['comb'] = data17['actor_1_name'] + ' ' + data17['actor_2_name'] + ' '+ data17['actor_3_name'] + ' '+ data17['director_name'] +' ' + data17['Genre']
data17
```
