```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
import pandas as pd
```

```python
d17 = pd.read_csv("/content/drive/MyDrive/Movie Datasets/data17.csv")
d18 = pd.read_csv("/content/drive/MyDrive/Movie Datasets/data18.csv")
d19 = pd.read_csv("/content/drive/MyDrive/Movie Datasets/data19.csv")
d20 = pd.read_csv("/content/drive/MyDrive/Movie Datasets/data20.csv")
d21 = pd.read_csv("/content/drive/MyDrive/Movie Datasets/data21.csv")
```

```python
d = pd.concat([d17,d18,d19,d20,d21],ignore_index = True)
d['Cast'] = d['actor_1_name'] + ' ' + d['actor_2_name'] + ' '+ d['actor_3_name'] + ' '+ d['director_name']
d['Title'] = d['Title'].str.lower()
d.rename(columns = {'Title':'movie_title'},inplace = True)
d
```

```python
d.to_csv('/content/drive/MyDrive/Movie Datasets/main_data.csv',index = False)
```
